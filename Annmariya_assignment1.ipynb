4.Maximum Finder

Task:

● Create a 4×4 random matrix.

● Find the maximum value.

● Find its position (row, column).

Hint:

First find the index in flattened form, then convert it back to row and column.

```python
arr=np.arange(5,21)
arr1=arr.reshape(4,4)
arr1
```

```python
max_value = np.max(arr1)
print("Maximum value:" ,max_value)
```

```python
max_index_flat = np.argmax(arr1)
max_position = np.unravel_index(max_index_flat, arr1.shape)
print("Position of maximum value (row, column):", max_position)
```

10.Simple Normalization

Task:

● Create a 3×3 random matrix.

● Divide each row by its row sum.

Hint:

Row sums need to keep their 2D shape so broadcasting works properly.

```python
import numpy as np

A = np.random.randint(1, 10, (3, 3))

row_sum = A.sum(axis=1, keepdims=True)

normalized = A / row_sum

print(normalized)
```
