1.A shopkeeper wants to calculate the total bill for two items. Write a program that takes
the price and quantity of each item and prints the total bill amount.


Sample Input:

Item1 price: 45

Item1 quantity: 3

Item2 price: 30

Item2 quantity: 2



```python
item1_price=int(input("enter the Item1_price "))
item2_price=int(input("enter the Item2_price "))
item1_quantity=int(input("enter theItem1_quantity "))
item2_quantity=int(input("enter theItem2_quantity "))
total_bill=(item1_price*item1_quantity)+(item2_price*item2_quantity)
print("Total bill is",total_bill)
```

8.A land surveyor wants to calculate the area of an irregular triangular plot.
Write a Python program to find the area of a triangle when all three sides are known,
using Heron’s formula:
𝑠 =
(𝑎 + 𝑏 + 𝑐)/
2

Area = √𝑠(𝑠 − 𝑎)(𝑠 − 𝑏)(𝑠 − 𝑐)

Sample Input:

Enter side a: 5

Enter side b: 6

Enter side c: 7

```python
import math
x=int(input("Enter side a "))
y=int(input("Enter side b "))
z=int(input("Enter side c "))
s=(x+y+z)/2
area=math.sqrt(s*(s-x)*(s-y)*(s-z))
print("Area of Triangle is",area)
```

9.A painter charges ₹25 per square meter. Write a program to calculate the total cost to
paint a rectangular wall. The wall has length and height, and there are two windows each
measuring 1.5m × 1.2m that should not be painted. Find the total area to be painted and
its total cost.

Sample Input:

Enter wall length (in meters): 12

Enter wall height (in meters): 8

```python
wall_length=int(input("Enter wall length (in meters) "))
wall_height=int(input("Enter wall height (in meters) "))
total_area=wall_length*wall_height
window_length=float(input("Enter window length (in meters) "))
window_height=float(input("Enter window height (in meters) "))
window_area=window_length*window_height
numberofwindows=2
notpaintedarea=window_area*numberofwindows
paintedarea=total_area-notpaintedarea
cost=paintedarea*25
print("Total Area to be painted is",paintedarea,"sq.m")
print("Total Cost is",cost)
```
