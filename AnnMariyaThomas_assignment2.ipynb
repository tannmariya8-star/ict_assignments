1.FizzBuzz: For a given range of numbers (e.g., 1 to 100), print "Fizz" if the number is
divisible by 3, "Buzz" if it's divisible by 5, and "FizzBuzz" if it's divisible by both. Otherwise,
print the number itself.

```python
j=int(input("enter the range"))
for i in range(1,j):
  if(i%3==0 and i%5==0):
    print("FizzBuzz")
  elif(i%3==0):
    print("Fizz")
  elif(i%5==0):
    print("Buzz")
```

2.Create a bill calculator where the user enters the number of units consumed. ₹5 per
unit for the first 100 units ₹7 per unit for the next 100 units (from 101 to 200) ₹10 per
unit for any units above 200 Calculate and display the total electricity bill.

Sample Input:

Enter units consumed: 245

Sample Output:

Total Bill Amount: ₹1650

```python
units = int(input("Enter units consumed: "))
total_bill = 0
if units <= 100:
    total_bill = units * 5
elif units <= 200:
    total_bill = (100 * 5) + ((units - 100) * 7)
else:
    total_bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)
print("Total Bill Amount: ₹",total_bill)
```

4.Develop a Python program that checks whether a given word or phrase is a
palindrome.

Sample input

Enter num or string to be checked = 121

ample Output

Given num/string is Palindrome

```python
s=input("Enter num or string to be checked = ")
s=str(s)  # Ensure input is treated as a string
reversed_s=s[::-1]
if s==reversed_s:
    print("Given num/string is Palindrome")
else:
    print("Given num/string is not Palindrome")
```
